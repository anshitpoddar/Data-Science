```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
X=[1,2,3]
Y=[1,3,2]
plt.scatter(X,Y);
```

```python
slopecurr=0
interceptcurr=0
LR = 0.01
itera = 1000
```

```python
for i in range(itera):
    sum1 = 0
    sum2 = 0
    for j in range(len(X)):
        sum1 += X[j] * (Y[j] - slopecurr*X[j] - interceptcurr)
        sum2 += Y[j] - slopecurr*X[j] - interceptcurr
    sgradient = -(2/len(X)) * sum1
    igradient = -(2/len(X)) * sum2
    slopecurr -= LR * sgradient
    interceptcurr -= LR * igradient
    y_predicted = [slopecurr*x + interceptcurr for x in X]
    plt.plot(X, y_predicted)
plt.scatter(X, Y)
y_predicted = [slopecurr*x + interceptcurr for x in X]
plt.plot(X, y_predicted)
plt.show()
```
